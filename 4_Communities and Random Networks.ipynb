<font size=4 face="黑体">ER graphs with 100 nodes</font>

```python
# compute the average degree for different edge creation probabilities p
#<K>=(N-1)p
k_list=[]
p_list=[]
for i in range(11):
    k_list.append(99*10**(-i))
    p_list.append(10**(-i))
```

```python
# generate ER graphs with N=100
# and compute NG/G, store it in l_list
# and compute the average clustering ⟨𝐶⟩ as a function of p, store it in c_list
l_list=[]
c_list=[]
for j in range (11):
    G = nx.erdos_renyi_graph(100,k_list[j])
    n=nx.number_connected_components(G)
    components = [c for c in nx.connected_components(G)]
    maxcomp = max(components, key=len)
    l=len(maxcomp)/100
    l_list.append(l)
    c=nx.average_clustering(G)
    c_list.append(c)
```

```python
# Plot the probability that a node belongs to the largest connected component 𝑁𝐺/𝑁 as a function of p
plt.plot(p_list,l_list)
plt.xscale('log')
plt.axvline(x=0.01,c="red")
plt.xlabel("edge creation probability p(log type)")
plt.ylabel("the probability N_G/N")
plt.title("relationship between the largest connected component and the edge creation probability")
plt.show()
```

```python
# Plot the average clustering ⟨𝐶⟩ as a function of p
plt.plot(p_list,c_list)
plt.xscale('log')
plt.xlabel("edge creation probability p(log type)")
plt.ylabel("the average clustering")
plt.title("relationship between the average clustering and the edge creation probability")
plt.show()
```

the result and the interpretation of it
(1)The average clustering coefficient increase as p increase.
Because when the possibility of forming an edge between nodes increase, the number of edges increase, so the possibility to form a triangle between three nodes increase. As a result, the average clustering coefficient increase.
(2)The average clustering coefficient change largely around p=0.01. 
Because when p is around 1/N (N=100), the average degree is around 1. So while increasing p, the possibility of forming a triangle increase largely.

<font size=6 face="黑体">A04.3 Small-world with high clustering</font>

```python
p_list=[]
for k in range(11):
    p_list.append(10**(-k))
print(p_list)
```

```python
# generate WS small-world networks
# and compute the average clustering for each p
cc_list=[]
for k in range(11):
    W = nx.watts_strogatz_graph(100,10,p_list[k])
    cc=nx.average_clustering(W)
    cc_list.append(cc)
print(cc_list)
```

```python
# generate the WS small-world network with p=0
# compute its average clustering ⟨𝐶(0)⟩
S=nx.watts_strogatz_graph(100,10,0)
cs=nx.average_clustering(S)
cs
```

```python
# compute ⟨𝐶(𝑝)⟩/⟨𝐶(0)⟩
cp0_list=[]
for k in range(11):
    cp=cc_list[k]/cs
    cp0_list.append(cp)
print(cp0_list)
```

```python
plt.plot(p_list,cp0_list)
plt.xscale('log')
plt.xlabel("the rewiring probability p")
plt.ylabel("the average clustering C(p)/C(0)")
plt.show()
```

```python
# compute the analytical result
k1=5
fp_list=[]
for k in range(11):
    fp=(3*(k1-1)*(1-p_list[k])**(3))/(2*(2*k1-1))
    fp_list.append(fp)
print(fp_list)
```

```python
# plot a graph of ⟨𝐶(𝑝)⟩ v.s. analytical result of ⟨𝐶(𝑝)⟩
plt.plot(cc_list,fp_list)
plt.xlabel("average clustering")
plt.ylabel("analytical result of average clustering")
plt.show()
```

```python
# as can be seen in the graph, average clustering and its analytical result are nearly the same
```

```python
# computethe average shortest-path length 𝐷(𝑝) for each p
D_list=[]
for k in range(11):
    W = nx.watts_strogatz_graph(100,10,p_list[k])
    g=nx.average_shortest_path_length(W) 
    D_list.append(g)
print(D_list)
```

```python
# computethe average shortest-path length 𝐷(0)
Ds = nx.average_shortest_path_length(S)
Ds
```

```python
# compute 𝐷(𝑝)/𝐷(0)
d_list=[]
for k in range(11):
    d=D_list[k]/Ds
    d_list.append(d)
print(d_list)
```

```python
# Plot the average shortest-path length 𝐷(𝑝)/𝐷(0)
plt.plot(p_list,d_list)
plt.xscale('log')
plt.xlabel("the rewiring probability p")
plt.ylabel("the average shortest-path length D(p)/D(0)")
plt.show()
```
